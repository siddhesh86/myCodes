```python
import os, sys
import numpy as np
from collections import OrderedDict as OD
import math
import hist
```

```python
h1DSimple = hist.Hist(hist.axis.Regular(5,0,5, name='x'), storage=hist.storage.Double() )
#h1DSimple
print(f"h1DSimple ({type(h1DSimple)}): \n{h1DSimple}, \n{h1DSimple.values() = } \n{h1DSimple.variances() = }\n")

h1DSimple[2] = 2
print(f"h1DSimple ({type(h1DSimple)}): \n{h1DSimple}, \n{h1DSimple.values() = } \n{h1DSimple.variances() = }\n")

h1DSimple[:] = np.random.uniform(low=1, high=10, size=5)
print(f"h1DSimple ({type(h1DSimple)}): \n{h1DSimple}, \n{h1DSimple.values() = } \n{h1DSimple.variances() = }\n")
```

```python
h1DWgted = hist.Hist(hist.axis.Regular(5,0,5, name='x'), storage=hist.storage.Weight() )
#h1DWgted
print(f"h1DWgted ({type(h1DWgted)}): \n{h1DWgted}, \n{h1DWgted.values() = } \n{h1DWgted.variances() = }\n")

h1DWgted[2] = [2, 1] 
print(f"h1DWgted ({type(h1DWgted)}): \n{h1DWgted}, \n{h1DWgted.values() = } \n{h1DWgted.variances() = }\n")

val_var = np.stack((np.random.uniform(low=1, high=10, size=5), np.random.uniform(low=0, high=1, size=5)), axis=-1)
print(f"val_var ({type(val_var)}) ({val_var.shape}): {val_var}")
h1DWgted[:] = val_var
print(f"h1DWgted ({type(h1DWgted)}): \n{h1DWgted}, \n{h1DWgted.values() = } \n{h1DWgted.variances() = }\n")
```

```python
h2DSimple = hist.Hist(hist.axis.Regular(5,0,5, name='x'), hist.axis.Regular(5,10,15, name='y'), storage=hist.storage.Double())
print(f"h2DSimple ({type(h2DSimple)}): \n{h2DSimple}, \n{h2DSimple.values() = } \n{h2DSimple.variances() = }\n")

h2DSimple[2, 3] = 2
print(f"h2DSimple ({type(h2DSimple)}): \n{h2DSimple}, \n{h2DSimple.values() = } \n{h2DSimple.variances() = }\n")

val_2d = np.arange(5*5).reshape((5,5))
print(f"val_2d ({type(val_2d)}): \n{val_2d}")
h2DSimple[:, :] = val_2d
print(f"h2DSimple ({type(h2DSimple)}): \n{h2DSimple}, \n{h2DSimple.values() = } \n{h2DSimple.variances() = }\n")
```

```python
h2DWgted = hist.Hist(hist.axis.Regular(5,0,5, name='x'), hist.axis.Regular(5,10,15, name='y'), storage=hist.storage.Weight() )
#h1DWgted
print(f"h2DWgted ({type(h2DWgted)}): \n{h2DWgted}, \n{h2DWgted.values() = } \n{h2DWgted.variances() = }\n")

h2DWgted[2, 3] = [2, 1]
print(f"h2DWgted ({type(h2DWgted)}): \n{h2DWgted}, \n{h2DWgted.values() = } \n{h2DWgted.variances() = }\n")

val_2d = np.arange(5*5).reshape((5,5))
var_2d = np.reshape(np.arange(5*5)/100, (5,5))
val_var_2d = np.stack((val_2d, var_2d), axis=-1)
print(f"val_2d ({type(val_2d)}) ({val_2d.shape}): \n{val_2d}")
print(f"var_2d ({type(var_2d)}) ({var_2d.shape}): \n{var_2d}")
print(f"val_var_2d ({type(val_var_2d)}) ({val_var_2d.shape}): \n{val_var_2d}")
h2DWgted[:, :] = val_var_2d
print(f"h2DWgted ({type(h2DWgted)}): \n{h2DWgted}, \n{h2DWgted.values() = } \n{h2DWgted.variances() = }\n")
```
